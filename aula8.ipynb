```python
#Criando nosso DataFrame
df= pd.read_Excel("adventureWorks.xlsx")
```

```python
#quantidade de linha e colunas
df.shape
```

```python
#verificar os tipos de dados
df.dtypes
```

```python
#qual a receita total
df["valor venda"].sum()
```

```python
#qual o custo total
df["custo"]=df["custo unitario"].mul(df["quantidade"])
```

```python
df.head(1)
```

```python
#qual o custo total
round(df["custo"].sum(),2)
```

```python
#Agora que temos a receita e custo toal, podemos achar o lucro total
#vamos criar uma coluna de lucro que será Receita - Custo
df["lucro"]= df["valor venda"]-df["custo"]
```

```python
df.head(1)
```

```python
#total lucro
round(df["lucro"].sum(),2)
```

```python
#criando uma coluna com total de dias para enviar o produto
df["tempo_envio"]=df["data envio"]- df["data venda"].dt.days
```

```python
df["tempo_envio"].head()
```

```python
#verificando o tipo
df["tempo_envio"].dtype
```

```python
df.groupby("marca")["tempo_envio"].mean()
```

```python
#gráfico total de produtos vendidos
df.gruopby("produto")["quantidade"].sum().sort_values(ascending=True).plot.hbar(title="Total Produtos Vendidos")
plt.xlabel("total")
plt.xlabel("Produto:");
```

```python
df.groupby(df["data venda"].dt.year)["lucro"].sum()
```

```python
#selecionar apenas as vendas de 2009
df_2009= df[df["data venda"].dt.year==2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df_2009["data venda"].dt.month)["lucro"].sum().plot(title="lucro x Mês")
plt.xlabel("mes")
plt.xlabel("lucro");
```

```python
df_2009.groupby("marca")["lucro"].sum().plot.bar(title="lucro x Marca")
plt.xlabel("marca")
plt.xlabel("lucro")
plt.xticks(rotation='horizontal');
```

```python
df_2009.groupby("classe")["lucro"].sum().plot.bar(title="lucro x classe")
plt.xlabel("classe")
plt.xlabel("lucro")
plt.xticks(rotation='horizontal');
```
